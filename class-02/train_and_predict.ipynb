```python
import numpy as np
from sklearn.datasets import load_iris

from perceptron import Perceptron, step_function
from trainer import Trainer
```

```python
X, y = load_iris(return_X_y=True)

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
np.random.seed(42)

y_setos_vs_rest = np.where(y == 0, 1, 0)

perceptron_setos_vs_rest = Perceptron(input_dim=X.shape[1], 
                                      activation_function=step_function)
trainer_setos_vs_rest = Trainer(epochs=128, 
                                learning_rate=0.001)

trainer_setos_vs_rest.train(X=X,
                            y=y_setos_vs_rest, 
                            perceptron=perceptron_setos_vs_rest)
trainer_setos_vs_rest.predict(X=X, 
                              y=y_setos_vs_rest, 
                              perceptron=perceptron_setos_vs_rest)
```

```python
np.random.seed(42)

y_veriscolour_vs_rest = np.where(y == 1, 1, 0)

perceptron_veriscolour_vs_rest = Perceptron(input_dim=X.shape[1], 
                                            activation_function=step_function)
trainer_veriscolour_vs_rest = Trainer(epochs=128, 
                                      learning_rate=0.001)

trainer_veriscolour_vs_rest.train(X=X,
                                  y=y_veriscolour_vs_rest, 
                                  perceptron=perceptron_veriscolour_vs_rest)
trainer_veriscolour_vs_rest.predict(X=X, 
                                    y=y_veriscolour_vs_rest, 
                                    perceptron=perceptron_veriscolour_vs_rest)
```

```python
np.random.seed(42)

y_virginica_vs_rest = np.where(y == 2, 1, 0)

perceptron_virginica_vs_rest = Perceptron(input_dim=X.shape[1], 
                                            activation_function=step_function)
trainer_virginica_vs_rest = Trainer(epochs=128, 
                                    learning_rate=0.001)

trainer_virginica_vs_rest.train(X=X,
                                y=y_virginica_vs_rest, 
                                perceptron=perceptron_veriscolour_vs_rest)
trainer_virginica_vs_rest.predict(X=X, 
                                  y=y_virginica_vs_rest, 
                                  perceptron=perceptron_veriscolour_vs_rest)
```
